```python
import requests
from bs4 import BeautifulSoup

url = "https://www.okdam.com"
response = requests.get(url)
bs = BeautifulSoup(response.content,'html.parser')
raw_menu = bs.find("ul",attrs={'class':'menu'}).find_all("li")
print(raw_menu)
```

```python
urls=[]
for category in raw_menu:
  c=category.find('a')['href']
  urls.append(c)
  
```

```python
urls
```

```python
res=requests.get(urls[0])
bs=BeautifulSoup(res.content,"html.parser")
```

```python
content=bs.find('div',attrs={'class':'feature-wrap clearfix'})
```

```python
products=content.find_all("div",attrs={'class':'product-box'})
```

```python
products
```

```python
heading = ['title','price','previous price']
data = []
for product in products:
  name=product.find("div",attrs={'class':'product_name-cat'})
  #price=product.find("span",{'class':'og-price'})
  price = product.find("span",{'class':'og-price'})
  #print(name.text,price.text)
  pprice=product.find('del')
  if pprice is not None:
    pprice=pprice.text
  else:
    pprice='na'
  data.append([name.text,price.text,pprice])
```

```python
data
```

```python
import csv
with open("okdam.csv","w") as f:
  writer = csv.writer(f)
  writer.writerow(['title','price','pprice'])
  writer.writerows(data)
```
